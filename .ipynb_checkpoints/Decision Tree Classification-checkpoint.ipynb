```python
from sklearn import tree
import pandas as pd
```

```python
data = pd.read_csv('../data/iris_with_answers.csv')
data['species_p'] = data['species'].replace(['setosa', 'versicolor', 'virginica'],[0,1,2])
```

```python
data
```

```python
x = data.iloc[:,0:4]
y = data['species_p']
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
yHat = clf.predict_proba(x)
```

```python
import matplotlib.pyplot as plt

features = data.columns.tolist()[0:4]
classes = data['species'].unique()

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=features,  
                   class_names=classes,
                   filled=True)
```
